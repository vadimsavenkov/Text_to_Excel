```python
list_of_files = glob.glob('Deviation/*.dev')

with open("wellname.csv", "w") as f:
    for fileName in list_of_files:
        data_list = open( fileName, "r" ).readlines()
        f.writelines(data_list[1:2])
df = pd.read_csv("wellname.csv", header = None, names=["#WELL NAME"])
df['#WELL NAME'] = df['#WELL NAME'].str.split(':').str[1]

df.head(3)
```

```python
with open("xcoord.csv", "w") as f:
    for fileName in list_of_files:
        data_list = open( fileName, "r" ).readlines()
        f.writelines(data_list[2:3])

df2 = pd.read_csv("xcoord.csv", header = None, names=["#WELL HEAD X-COORDINATE, m"]);
df2['#WELL HEAD X-COORDINATE, m'] = df2['#WELL HEAD X-COORDINATE, m'].str.split(':').str[1].str.split('(').str[0]
df2.head(3)
```

```python
with open("ycoord.csv", "w") as f:
    for fileName in list_of_files:
        data_list = open( fileName, "r" ).readlines()
        f.writelines(data_list[3:4])

df3 = pd.read_csv("ycoord.csv", header = None, names=["#WELL HEAD Y-COORDINATE, m"]);
df3['#WELL HEAD Y-COORDINATE, m'] = df3['#WELL HEAD Y-COORDINATE, m'].str.split(':').str[1].str.split('(').str[0]
df3.head(3)
```

```python
list_of_files = glob.glob('Deviation/*.dev')

with open("offset.csv", "w") as f:
    for fileName in list_of_files:
        data_list = open( fileName, "r" ).readlines()
        f.writelines(data_list[4:5])

df4 = pd.read_csv("offset.csv", header = None, names=["# WELL OFFSET (from MSL), ft"]);
df4['# WELL OFFSET (from MSL), ft'] = df4['# WELL OFFSET (from MSL), ft'].str.split(':').str[1].str.split('(').str[0]
df4.head(3)
```

```python
df_combined = pd.concat([df, df2, df3, df4], axis=1)
df_combined.to_excel("table.xlsx", index=False)
df_combined
```
